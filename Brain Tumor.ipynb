```python
################# Outliers ##################################################
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1

index_outliers=df.loc[((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)].index.tolist()
list_outliers=df.loc[index_outliers]

remove_outliers=df[~((df<(Q1-1.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)]

df_final=remove_outliers
```

```python
X = df_final.iloc[:,1:7].values
y = df_final.iloc[:,9].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_reg = DecisionTreeClassifier()
tree_reg.fit(X_train,y_train)

tree_reg_r2_train=tree_reg.score(X_train,y_train)
print(tree_reg_r2_train)
tree_reg_r2_test=tree_reg.score(X_test,y_test)
print(tree_reg_r2_test)

# Predicting the Test set results
y_pred = tree_reg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
#fetching accuracy
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print('Accuracy Score :',accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
tree_reg = RandomForestClassifier()
tree_reg.fit(X_train,y_train)

tree_reg_r2_train=tree_reg.score(X_train,y_train)
print(tree_reg_r2_train)
tree_reg_r2_test=tree_reg.score(X_test,y_test)
print(tree_reg_r2_test)

# Predicting the Test set results
y_pred = tree_reg.predict(X_test)
```

```python
from xgboost import XGBClassifier
tree_reg = XGBClassifier()
tree_reg.fit(X_train,y_train)

tree_reg_r2_train=tree_reg.score(X_train,y_train)
print(tree_reg_r2_train)
tree_reg_r2_test=tree_reg.score(X_test,y_test)
print(tree_reg_r2_test)
```
